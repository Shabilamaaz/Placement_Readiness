```python
import pandas as pd
import sqlite3

conn = sqlite3.connect("database/database.db")
data = pd.read_sql("SELECT * FROM students", conn)

data.head()
```

# CLEANING

```python
data['internships'] = data['internships'].replace({'Yes':1, 'No':0})
data['placed'] = data['placed'].replace({'Yes':1, 'No':0})
data['projects_count'] = data['projects_count'].replace({'Yes':1, 'No':0})

data['projects_count'] = pd.to_numeric(data['projects_count'])
```

```python
data_clean = data.drop(['student_id','student_name','course','batch'], axis=1)
```

```python
data_clean.info()
```

```python
data_clean = data_clean.dropna(subset=['placed'])
```

```python
data_clean.info()
```

```python
X = data_clean.drop('placed', axis=1)
y = data_clean['placed']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

# Prediction

```python
y_pred = model.predict(X_test)

print(y_pred[:10])
```

# Accuracy

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, y_pred)
print("Accuracy:", acc)
```

```python
new_student = [[6, 8.5, 85, 75, 80, 78, 3, 1]]
```

```python
result = model.predict(new_student)

if result[0] == 1:
    print("Placed ✅")
else:
    print("Not Placed ❌")
```

```python
import pickle

# Load model
with open('model/model.pkl', 'rb') as f:
    model = pickle.load(f)

# Prediction function
def predict_placement(data):
    result = model.predict([data])
    
    if result[0] == 1:
        return "Placed ✅"
    else:
        return "Not Placed ❌"


# Test (optional)
if __name__ == "__main__":
    sample = [6, 8.5, 85, 75, 80, 78, 3, 1]
    print(predict_placement(sample))
```
